# Gender

Melihat distribusi data gender yang dibagi kedalam dua kategori
M = Male
F = Female

kedalam bentuk barplot dan pieplot

```python
df['gender'].value_counts()
```

```python
plt.title('distribusi gender')
sns.countplot(x=df['gender'])
plt.show()
```

```python
plt.title('distribusi gender')
plt.pie(x= df['gender'].value_counts().values, labels=df['gender'].value_counts().index, autopct='%1.f%%')
plt.show()
```

```python
fig ,(ax1,ax2) = plt.subplots(1,2, figsize=(9,5))

plt.suptitle('Distribusi Gender')

ax1.bar(x=df.gender.value_counts().index, height = df.gender.value_counts().values, color = ['r','g'])

ax2.pie(x=df.gender.value_counts().values, labels=df.gender.value_counts().index, autopct='%1.1f%%', colors= ['r','g'])

legend_handle= [
    mpatches.Patch(color='r', label= 'Laki-Laki'),
    mpatches.Patch(color='g', label='Perempuan')
]
fig.legend(handles=legend_handle, title='Jenis Kelamin')
plt.show()
```

Dapat ditarik kesimpulan :
- distribusi data laki-laki dalam dataset lebih banyak daripada distribusi data perempuan di fitur gender

# Checking Korelasi antara Fitur Gender dan Final_result

```python
final_gender =df.groupby(['gender','final_result'])['gender'].value_counts()
final_gender
```

```python
plt.title('Final result berdasarkan gender')
sns.barplot(x=final_gender.index.get_level_values('final_result'), y=final_gender.values, hue= final_gender.index.get_level_values('gender'))
plt.show()
```

insight yang didapat dari analisis hasil visualisasi diatas adalah :
1. perempuan lebih banyak distinction, perempuan lebih mungkin untuk mencapai hasil tertinggi.
2. Data tersebut menunjukkan bahwa lebih banyak individu laki-laki yang mengalami kegagalan ketimbang perempuan
3.  Data ini juga memungkinkan kita untuk membandingkan persentase perempuan dan laki-laki yang lulus atau mengundurkan diri dari program. Ini bisa memberikan wawasan tentang perbedaan dalam tingkat kelulusan atau mengundurkan diri antara jenis kelamin.

# Checking banyaknya individu disabilitas tiap gender

```python
final_gender_disability = df.groupby(['gender','disability'])['gender'].value_counts()
final_gender_disability
```

1. jumlah individu yang tidak memiliki disabilitas jauh lebih besar daripada jumlah individu yang memiliki disabilitas.
2. walaupun distribusi gender laki-laki jauh diatas perempuan. namun dapat dilihat justru lebih banyak individu dengan gender perempuan lah yang lebih banyak disabilitas

```python
fig, (axs1, axs2) = plt.subplots(1,2, figsize=(10,5))
filtered_data = final_gender_disability.loc[final_gender_disability.index.get_level_values('gender') == 'F']
filtered_data_2 = final_gender_disability.loc[final_gender_disability.index.get_level_values('gender') == 'M']
plt.suptitle('Persentase disabilitas tiap gender')

axs1.set_title('Female')
axs1.pie(x=filtered_data.values, labels=filtered_data.index.get_level_values('disability'), autopct='%1.1f%%')

axs2.set_title('Male')
axs2.pie(x=filtered_data_2.values, labels=filtered_data_2.index.get_level_values('disability'), autopct='%1.1f%%')

legend_hndle = [
    mpatches.Patch(label= 'Normal'),
    mpatches.Patch(label= 'Disabilitas', color = 'orange') 
]

fig.legend(handles=legend_hndle, title='kategori')

plt.show()
```

```python
sns.barplot(
    x=final_gender_disability.index.get_level_values('gender'), y= final_gender_disability.values, hue= final_gender_disability.index.get_level_values('disability')
)
plt.show()
```
